# normalization and Z-normalization

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("data/after_3s.csv")
df.drop(columns =['Unnamed: 0'], axis = 1, inplace = True)
df.head()
```

```python
# исследуем разброс диапазона данных
fig, ax = plt.subplots(figsize = (16, 6))
df.plot(kind = 'kde', ax = ax);
```

# ТУТ лучше опять сделать функцию и отрисовать каждый признак по отдельности

# Normalization

```python
min_max_scaler = preprocessing.MinMaxScaler()
df_norm = pd.DataFrame(min_max_scaler.fit_transform(df), columns = df.columns, index = df.index)
df_norm
```

```python
for i in df_norm.columns:
    plt.figure(figsize = (8,4))
    plt.title('Гистограмма: ' + ' ' + i)
    plt.ylabel('Количество элементов')
    sns.histplot(data = df_norm[i], kde = True)
    plt.savefig('data/Гистограмма после норм данных.pdf')# только последнюю?
    plt.show()
    print(f'Минимальное значение: {df_norm[i].min()}')
    print(f'Максимальное Значение: {df_norm[i].max()}')
    print(f'Среднее значение: {df_norm[i].mean()}')
    print(f'Медианное значение: {df_norm[i].median()}')
```

# ПОЧЕМУ ДАННЫЕ В ГИСТОГРАММЕ СТРАННЫЕ?
# должны быть от 0 до 1....(вроде убрал)

```python
df_norm.describe().round(3).T
```

```python
boxplot = df_norm.boxplot(rot = 90, figsize = (16,6))
```

сохраним нормализованные данные

```python
df_norm.to_csv('data/df_norm.csv')
```

# Z-normalization(стандартизация)

```python
scaler = StandardScaler()
df_stand = pd.DataFrame(scaler.fit_transform(df), columns = df.columns, index = df.index)
df_stand.head()
```

```python
for i in df_stand.columns:
    plt.figure(figsize = (8,4))
    plt.title('Гистограмма: ' + ' ' + i)
    plt.ylabel('Количество элементов')
    sns.histplot(data = df_stand[i], kde = True)
    plt.savefig('data/Гистограмма после стданд данных.pdf')# только последнюю?
    plt.show()
    print(f'Минимальное значение: {df_stand[i].min()}')
    print(f'Максимальное Значение: {df_stand[i].max()}')
    print(f'Среднее значение: {df_stand[i].mean()}')
    print(f'Медианное значение: {df_stand[i].median()}')
```

```python
df_stand.describe().round(3).T
```

```python
boxplot = df_stand.boxplot(rot = 90, figsize = (16,6))
```

сохраним стандартизированные данные

```python
df_stand.to_csv('data/df_stand.csv')
```
