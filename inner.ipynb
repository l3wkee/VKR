# Обьеденяем с помощью INNER

```python
df = df1.merge(df2, left_index = True, right_index = True, how = 'inner')
```

```python
df.head()
```

удаляем ненужные столбцы(колонки)

```python
df.drop(columns =['Unnamed: 0_x', 'Unnamed: 0_y'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
df.info()
```

сохраним датафрейм

```python
df.to_csv('data/after_inner.csv')
```
